```python
import json
import pandas as pd
import numpy as np
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
from __future__ import print_function
from collections import defaultdict
```

```python
with open('ingrds.txt') as f:
    content = f.readlines()
    
ingredient_list = [a.strip() for a in content]

ing_lookup = defaultdict(list)
for ing in ingredient_list:
#     print(ing)
    ing_lookup[ing.lower()[0]].append(ing.lower())
    
display(ing_lookup)
```

```python
final_recipes = []

import os
names = []
for file in os.listdir("new_data/"):
    if file.endswith(".json"):
        names.append(file)

for name in names:
    print("\n=== File {} ===".format(name))
    json_data = open('new_data/'+name)
    recipes = json.load(json_data)

    for idx, recipe in enumerate(recipes):
        ingredients = []
        for val in recipe['ingredients']:
            for token in val.split(" "):
                token = token.strip(",").lower()
                if token != "" and token.isdigit() is False:
                    if token in ing_lookup[token[0]]:
                        new_ing = {}
                        new_ing['name'] = token
                        new_ing['unit'] = None
                        new_ing['q'] = None
                        ingredients.append(new_ing)
#                     result = process.extractOne(token, ing_lookup[token[0]])
#                     if result:
#                         if result[1] > 90:
#                             new_ing = {}
#                             new_ing['name'] = result[0]
#                             new_ing['unit'] = None
#                             new_ing['q'] = None
#                             ingredients.append(new_ing)
#                                 print(token, " | ", result[0])
                        break
        recipe['ingredients'] = ingredients
        final_recipes.append(recipe)
```

```python
columns = [u'Name', u'Rating']
columns.extend(ingredient_list)

frame = pd.DataFrame(columns = columns)

for idx, recipe in enumerate(final_recipes):
    if recipe['ingredients']:
        for ingredient in recipe['ingredients']:
            ings = [1 if ingredient['name'] == list_element else 0 for list_element in ingredient_list]
    else:
        ings = [0 for list_element in ingredient_list]
    row = [recipe['name'], recipe['rating']]
    row.extend(ings)
    frame.loc[idx] = row
    
display(frame)
```

```python
df = frame.copy()

df.Rating = df.Rating.astype(float)
df.Rating = df.Rating.apply(lambda x: x//((df.Rating.max()-df.Rating.min())/2))
df.Rating.describe()
```

```python
cols = df.columns.values[2:-1]
training = df[cols].values
print(training)
```

```python
from sklearn import tree
from sklearn.cross_validation import KFold
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier

accs = []

X = training
y = df.Rating.values

kf = KFold(X.shape[0], n_folds=5, shuffle=True)
clf = RandomForestClassifier(n_estimators=200, max_depth=None,random_state=0)

for train_index, test_index in kf:
#     print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
#     clf = tree.DecisionTreeClassifier()
    preds = clf.fit(X_train, y_train).predict(X_test)
    
    accs.append(accuracy_score(y_test, preds))
    
print(accs)
```
